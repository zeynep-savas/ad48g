#Loaded the data.table package.

```r
library(data.table)
library(tibble)
library(tidyverse)
library(ggplot2)
library(GGally)
```

#Set the data as tibble.

```r
data("nasa", package = "GGally")
nasa_data <- as.tibble(nasa)
```

#Selected "id", "date", and "pressure" columns.

```r
nasa_data %>%
select(id, date, pressure)
```

#Filtered rows for columns when "cloudlow" is above 40 and "pressure" is below 1000.

```r
nasa_data %>% 
filter(cloudlow > 40 & pressure < 1000)
```

#Created new columns "high_low" and "o_p".

```r
nasa_data %<>%
mutate( high_low = cloudhigh / cloudlow ) %>%
mutate( o_p = ozone * pressure)
```

#Grouped the tibble by "id" and found max values for each id.

```r
nasa_data %>%
group_by(id) %>%
summarize(max_temperature = max(temperature, na.rm = T))
```

#Executed a pivot_wider operation with "nasa_data".

```r
nasa_at <- nasa_data %>%
group_by(id, pressure) %>%
summarise(min_at_nasa = min(lat, na.rm = T))
```

```r
nasa_at
```

```r
nasa_at_wide <- nasa_at %>% pivot_wider(id_cols = id, names_from = pressure , values_from = min_at_nasa)
```

#Executed a pivot_longer operation with "nasa_at_wide".

```r
nasa_at_wide %>% pivot_longer(cols = -"id", names_to = "pressure2", values_to = "min_nasa2", values_drop_na = T)
```

#Created a new table for left_join operation by using slice().

```r
table2 <- nasa_data %>% slice(400:1200)
```

```r
#Made a left_join operation by "date".
```

```r
nasa_data %>% left_join(table2, by = "date")
```
